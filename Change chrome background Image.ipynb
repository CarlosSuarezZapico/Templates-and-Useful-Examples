```python
import numpy as np
import matplotlib.image as npimg
import matplotlib.pyplot as plt
import cv2
%matplotlib qt
```

```python
image2 = npimg.imread('croma.jpg')
image2_copy = np.copy(image2)
image2.shape
```

```python
plt.imshow(image2, cmap='gray')
```

```python
t = image2[(0,0)]
t
```

```python
#Define our color selection boundaries in RGB values
lower_blue = np.array([0,230,0])
upper_blue = np.array([100,255,100])                     
```

# Create a Mask

```python
#define the mark area
mask = cv2.inRange(image2, lower_blue, upper_blue)
plt.imshow(mask, cmap='gray')
```

```python
# Mask the image to let the woman show through
masked_image = np.copy(image2_copy)

masked_image[mask !=0] = [0,0,0]

plt.imshow(masked_image)
```

# Mask and add background image

```python
background_image = cv2.imread('mountain.jpg')
background_image= cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)
background_image.shape
```

```python
crop_background = background_image[2000:2000+480, 5000:5000+852]
crop_background[mask==0]=[0,0,0]
plt.imshow(crop_background)
```

```python
complete_image = crop_background + masked_image
plt.imshow(complete_image)
```
